```python
#importing the required libs

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the csv file containing data
dataset=pd.read_csv('salary.csv')
print(dataset.head())
```

```python
#Since we have only 1 feature column , separating feature and label into separate variables
X=dataset.iloc[:,:1].values
y=dataset.iloc[:,1].values
```

```python
#splitting into test and training set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

#Fitting simple linear regressor to training set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

#predicting the test result
print(X_test)
y_pred=regressor.predict(X_test)

#visualizing the training set results
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Linear Regressor Train')
plt.xlabel('Years_of_Exp')
plt.ylabel('Salary')
plt.show()

#visualizing the test set results
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Linear Regressor Test')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```
